```python
# Use the scraper and parser scripts (see src subfolder) to get clean fight data

import sys; sys.path.append("..")
import numpy as np

from src.scraper import build_master_fighter_list, scrape_all_fighters
from src.parse_fights import clean_fight_dataframe

fighters = build_master_fighter_list()
raw_fights = scrape_all_fighters(fighters)
clean_fights = clean_fight_dataframe(raw_fights)
```

```python
fights_final = clean_fights.copy()

#ensure strings
f = fights_final["fighter"].astype(str)
o = fights_final["opponent"].astype(str)

fights_final["pair_f"] = np.where(f <= o, f, o)
fights_final["pair_o"] = np.where(f <= o, o, f)

fights_final = fights_final.drop_duplicates(subset=["date", "pair_f", "pair_o"], keep="first")
fights_final = fights_final.drop(columns=["pair_f", "pair_o"])
```

```python
# save to raw/processed data folder
import os

raw_fights.to_csv("../data/raw/raw_fight_history.csv", index=False)
print("Saved to ../data/raw/raw_fight_history.csv")

os.makedirs("../data/processed", exist_ok=True)
fights_final.to_csv("../data/processed/fight_history.csv", index=False)
print("Saved to ../data/processed/fight_history.csv")

fighters.to_csv("../data/processed/fighter_list.csv", index=False)
print("Saved to ../data/processed/fighter_list.csv")
```
